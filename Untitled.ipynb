```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

print("TensorFlow Version:", tf.__version__)
```

```python
# --- Final Upgraded Code for Cell 4 ---
from tensorflow.keras.callbacks import EarlyStopping

# Load the dataset
try:
    df = pd.read_csv('data/ethereum_2015-08-07_2024-09-08.csv')
except FileNotFoundError:
    print("Error: Dataset file not found.")
    # Create dummy data if not found
    dates = pd.date_range(start='2015-08-07', end='2024-09-08', freq='D')
    price = np.linspace(1, 4000, len(dates)) + np.random.randn(len(dates)) * 50
    volume = np.random.randint(100000, 1000000, len(dates))
    df = pd.DataFrame({'Start': dates, 'Close': price, 'Volume': volume, 'High': price + 10, 'Low': price - 10})
    print("Using a dummy dataframe for demonstration.")


# Convert 'Start' to datetime and sort
df['Start'] = pd.to_datetime(df['Start'])
df = df.sort_values('Start').reset_index(drop=True)

# --- STATE-OF-THE-ART FEATURE ENGINEERING ---
# 1. RSI
delta = df['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
df['RSI'] = 100 - (100 / (1 + (gain / loss)))

# 2. MACD
exp1 = df['Close'].ewm(span=12, adjust=False).mean()
exp2 = df['Close'].ewm(span=26, adjust=False).mean()
df['MACD'] = exp1 - exp2

# 3. Average True Range (ATR) - Volatility
df['TR'] = np.maximum(df['High'] - df['Low'], np.maximum(abs(df['High'] - df['Close'].shift()), abs(df['Low'] - df['Close'].shift())))
df['ATR'] = df['TR'].rolling(window=14).mean()

# 4. On-Balance Volume (OBV) - Volume Momentum
df['OBV'] = (np.sign(df['Close'].diff()) * df['Volume']).fillna(0).cumsum()

# 5. High-Low Volatility
df['HLT'] = ((df['High'] - df['Low']) / df['Close']) * 100


# --- TARGET VARIABLE ENGINEERING: Percentage Change ---
df['pct_change'] = df['Close'].pct_change()

# --- DATA PREPARATION ---
features = ['Close', 'Volume', 'RSI', 'MACD', 'ATR', 'OBV', 'HLT']
target = 'pct_change'

# Drop all rows with any NaN values
full_df = df[features + [target] + ['Start']].dropna().reset_index(drop=True)

# Separate features (X) and target (y)
X_data = full_df[features]
y_data = full_df[target]
dates = full_df['Start'] # Keep aligned dates for plotting

print("Final Features:")
print(X_data.head())

# --- DATA SCALING & SPLITTING ---
train_size = int(len(X_data) * 0.8)
X_train_df, X_test_df = X_data[:train_size], X_data[train_size:]
y_train_series, y_test_series = y_data[:train_size], y_data[train_size:]

# Scale the feature data
scaler = MinMaxScaler(feature_range=(0, 1))
X_train_scaled = scaler.fit_transform(X_train_df)
X_test_scaled = scaler.transform(X_test_df)

# Convert target series to numpy array
y_train, y_test = y_train_series.values, y_test_series.values
```

```python
# --- Code for Cell 6 (Ensure it uses new variables) ---
def create_sequences(X_data, y_data, look_back=30):
    X, y = [], []
    for i in range(look_back, len(X_data)):
        X.append(X_data[i-look_back:i, :])
        y.append(y_data[i])
    return np.array(X), np.array(y)

look_back = 30
X_train, y_train = create_sequences(X_train_scaled, y_train_scaled, look_back)
X_test, y_test = create_sequences(X_test_scaled, y_test_scaled, look_back)

# Reshape y_test for later calculations
y_test_unscaled = y_test_series.iloc[look_back:].values

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test_unscaled shape: {y_test_unscaled.shape}")
```

```python
# --- Corrected Code for Cell 10 ---
from tensorflow.keras.layers import GRU

model = Sequential()
num_features = X_train.shape[2]

# Layer 1: Larger GRU to process rich features
model.add(GRU(units=100, return_sequences=True, input_shape=(look_back, num_features)))
model.add(Dropout(0.2))

# Layer 2: Smaller GRU to consolidate
model.add(GRU(units=50))
model.add(Dropout(0.2))

# Dense Layer
model.add(Dense(units=25, activation='relu'))
# Output Layer predicts a single value (the percent change)
model.add(Dense(units=1))

model.summary()
```

```python
# --- Corrected Code for Cell 12 ---

# Check if a GPU is available
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))

model.compile(optimizer='adam', loss='mean_squared_error')

# Define the EarlyStopping callback
# It will monitor validation loss and stop if it doesn't improve for 10 epochs
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

print("Starting model training with Early Stopping...")
history = model.fit(
    X_train,
    y_train,
    epochs=200, # Max epochs, but will likely stop earlier
    batch_size=64, # Increased batch size for speed
    validation_split=0.1,
    callbacks=[early_stop], # Add the callback here
    verbose=1
)
print("Model training is complete.")
```

```python
# --- Final Upgraded Code for Evaluation Cell ---

# Make predictions (these are the predicted percent changes)
predicted_pct_change = model.predict(X_test)

# --- CONVERT PREDICTIONS BACK TO PRICE ---
# To get the predicted price, we need the actual price from the day before the prediction starts.
# We find the corresponding 'Close' prices from the original unscaled test dataframe.
previous_day_prices = X_test_df['Close'].iloc[look_back-1:-1].values

# Calculate the predicted price: Price_today = Price_yesterday * (1 + pct_change_today)
# We need to reshape for broadcasting
predicted_price = previous_day_prices.reshape(-1, 1) * (1 + predicted_pct_change)


# Get the actual prices for the same period
actual_price = X_test_df['Close'].iloc[look_back:].values

# --- CALCULATE DIRECTIONAL ACCURACY ---
results_df = pd.DataFrame({
    'Actual_Price': actual_price,
    'Predicted_Price': predicted_price.flatten(),
    'Previous_Actual_Price': previous_day_prices
})

# Determine actual and predicted direction relative to the PREVIOUS day's close
results_df['Actual_Direction'] = (results_df['Actual_Price'] > results_df['Previous_Actual_Price']).astype(int)
results_df['Predicted_Direction'] = (results_df['Predicted_Price'] > results_df['Previous_Actual_Price']).astype(int)

correct_predictions = (results_df['Actual_Direction'] == results_df['Predicted_Direction']).sum()
total_predictions = len(results_df)

directional_accuracy = (correct_predictions / total_predictions) * 100
print(f"\nFinal Directional Accuracy: {directional_accuracy:.2f}%")


# --- PLOTTING ---
# Get the correct dates for the plot
plot_dates = dates.iloc[train_size + look_back:]

plt.figure(figsize=(16, 8))
plt.title('Price Prediction vs. Actual (Upgraded Regression Model)', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.plot(plot_dates.values, actual_price, color='blue', label='Actual Price')
plt.plot(plot_dates.values, predicted_price, color='red', linestyle='-', label='Predicted Price')
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
